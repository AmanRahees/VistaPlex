Here genres, cast and crew are in list but it contains a dictionary.

```python
# convert the genre, cast and crew column into a list
import ast 
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x)) # "literal_eval" helps to convert the string into a list
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
# dictionary value take out in genres column
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_genresList(x))
```

```python
data['genres_list']
```

```python
# dictionary value take out in cast column
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

```python
data['actor_1_name']
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<= 1:
        return np.NaN
    else:
        return (casts[1])
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python
data['actor_2_name']
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<= 2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
data['actor_3_name']
```

```python
data['crew'][1]
```

```python
# Directors
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data['director_name']
```

```python
movie = data.loc[:, ['director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'genres_list', 'title']]
```
